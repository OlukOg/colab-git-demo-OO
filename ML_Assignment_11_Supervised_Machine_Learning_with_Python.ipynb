```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
# Load dataset
df = pd.read_csv('iris.csv')

# Load the first few rows of the dataset
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

No missing values and all features are numeric except the target (variety), which is categorical.



```python
# Encoding categorical data

le = LabelEncoder()
df['variety'] = le.fit_transform(df['variety'])

# Setosa = 0, Versicolor = 1, Virginica = 2

df.head()
```

The variety column contains string labels ("Setosa", etc.). LabelEncoder converts them into numerical labels required by ML models.

Task 4:Model Training

```python
from re import X
# Splitting the dataset into the Training set and Test set
X = df.drop(columns=['variety'])
y = df['variety']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

The independent features X (sepal/petal measurements) and the dependent target y (encoded flower type).
Datasets were Splitted:80% for training,20% for testing and random_state ensures reproducibility

```python
y_train.head()
```

```python
# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Used StandardScaler to standardize feature values (mean = 0, std = 1), which improves performance of distance-based models like SVM and KNN.

```python
X_train
```

```python
# Model Training: Using Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
accuracy = model.score(X_test, y_test)*100
print("Accuracy:", accuracy)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model training
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC()
}

results = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    results[name] = (accuracy, report)

for model, (accuracy, report) in results.items():
    print(f"{model} Accuracy: {accuracy:.2f}")
    print(report)
```

Three models—Logistic Regression, Random Forest, and SVM—were trained and evaluated using accuracy, precision, recall, and F1-score. All achieved perfect scores (accuracy = 1.00) due to clean data, clear class separation, and strong features. Logistic Regression alone sufficed, while SVM and Random Forest confirmed robustness. For more thorough validation, applying k-fold cross-validation, testing on noisy or real-world data, and visualizing decision boundaries are recommended.
